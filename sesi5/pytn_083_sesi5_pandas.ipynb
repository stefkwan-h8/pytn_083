# Query + Groupby + Aggregation
kita bisa mencari data dengan kondisi tertentu, kita group per kolom tertentu, dan kita agregasikan

```python
# tim Huskies (fran_id)
# di tahun 1947 (year_id)

# grup per tim Huskies lawan siapa (opp_fran)
# dan apakah dia menang atau kalah (game_result)

# agregasi menggunakan count() di kolom game_id untuk hitung total game

kondisi1 = df_nba["fran_id"] == "Huskies"
kondisi2 = df_nba["year_id"] == 1947

# df_huskies = df_nba[kondisi1 & kondisi2]
# df_huskies

df_nba[kondisi1 & kondisi2]["game_id"].count()
```

```python
df_nba[kondisi1 & kondisi2].groupby(["opp_fran", "game_result"])["game_id"].count()
```

di tahun 1947 (year_id) tim kita Huskies (fran_id) melawan siapa saja (opp_fran) dan menang/kalah berapa kali untuk setiap lawan
