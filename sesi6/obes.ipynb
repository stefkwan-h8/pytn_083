# Analisa Data
kita mau analisa data obesitas di England menggunakan file data obes.xls

```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import data
df_obes = pd.read_excel("obes.xls")
df_obes
```

```python
# cek file ini ada sheet apa saja menggunakan pandas
xc_obes = pd.ExcelFile("obes.xls")

print(xc_obes.sheet_names)
```

```python
# data bisa kita parse dari xc_obes 
# atau kita read langsung menggunakan read_excel 
# dengan tambahan opsi sheet ke berapa yang mau dibaca

df_obes72 = xc_obes.parse('7.2', skiprows=4, skipfooter=14)
df_obes72.drop(index=[0], inplace=True)
df_obes72
```

```python
df_obes72 = pd.read_excel("obes.xls", sheet_name="7.2", skiprows=4, skipfooter=14)
df_obes72.drop(index=[0], inplace=True)
df_obes72
```

```python
# rename column
df_obes72.rename(columns={"Unnamed: 0": "Date"}, inplace=True)
df_obes72
```

```python
# cek info dan isnull juga tetap bisa
df_obes72.info()
```

```python
# menggunakan kolom date untuk jadi index

df_obes72 = df_obes72.set_index("Date")
# df_obes72 = df_obes72.set_index("Date", inplace=True)
df_obes72
```

```python
# di line plot, setiap kolom di DataFrame akan menjadi 1 garis data, 1 set data
df_obes72.plot()
```

```python
# style sebuah plot bisa kita ganti
# list styles ada di https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
mpl.style.use(["ggplot"])

df_obes72.plot()
```

```python
# yang kolom Total mau kita exclude
df_obes72 = df_obes72.drop(columns=["Total"])
df_obes72.plot()
```

```python
df_obes72.plot(kind="bar")
```

```python
df_obes72.plot(kind="barh")
```

```python
df_obes72.plot()

plt.title("Obesity in England")
plt.xlabel("Year")
plt.ylabel("Total Number of People")
```
